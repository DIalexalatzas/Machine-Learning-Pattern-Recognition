## **1.1**
Έστω μια συνάρτηση $f: \mathbb{R}^d \rightarrow \mathbb{R}$, με $f(\mathbf{x}) = (\mathbf{x} - \mathbf{a})^T \mathbf{A} (\mathbf{x} - \mathbf{a})$, όπου $ \mathbf{A} \in \mathbb{R}^{d \times d}$ είναι ένας συμμετρικός πίνακας και  $\mathbf{a} \in \mathbb{R}^d$. Να υπολογίσετε αναλυτικά, βήμα προς βήμα με το χέρι, την παράγωγο $\nabla f$ της $f$ (ως προς το διάνυσμα $\mathbf{x} \in \mathbb{R}^{d}$) παραθέτοντας όλα τα ενδιάμεσα βήματα.


Αρχικά θα έχουμε ότι:
$f(\mathbf{x}) = (\mathbf{x} - \mathbf{a})^T \mathbf{A} (\mathbf{x} - \mathbf{a})) = \mathbf{x}^T\mathbf{A}\mathbf{x} - \mathbf{x}^T\mathbf{A}\mathbf{a} - \mathbf{a}^T\mathbf{A}\mathbf{x} + \mathbf{a}^T\mathbf{A}\mathbf{a}$. \\
Και θα υπολογίσουμε τις παραγώγους των στοιχείων ως προς $\mathbf{x}$. \\


*   $\frac{\partial}{\partial \mathbf{x}} (\mathbf{x}^T \mathbf{A} \mathbf{x}) = (\mathbf{A} + \mathbf{A}^T) \mathbf{x}$
*   $\frac{\partial}{\partial \mathbf{x}} (-\mathbf{x}^T \mathbf{A} \mathbf{a}) = -\mathbf{A}^T \mathbf{a}$
*   $\frac{\partial}{\partial \mathbf{x}} (-\mathbf{a}^T \mathbf{A} \mathbf{x}) = -\mathbf{A} \mathbf{a}$
*   $\frac{\partial}{\partial \mathbf{x}} (\mathbf{a}^T \mathbf{A} \mathbf{a}) = 0$

Προσθέτοντας όλους τους όρους, μας δίνεται η παράγωγος $\nabla f(\mathbf{x}) = (\mathbf{A} + \mathbf{A}^T) \mathbf{x} - \mathbf{A}^T \mathbf{a} -\mathbf{A} \mathbf{a}$

Επειδή μας δίνεται ότι ο $\mathbf{A}$ είναι συμμετρικός πίνακας, δηλαδή $\mathbf{A}^T = \mathbf{A}$, τελικά έχουμε ότι:
$\nabla f(\mathbf{x}) = 2\mathbf{A}\mathbf{x} - 2\mathbf{A}\mathbf{a} = 2\mathbf{A}(\mathbf{x}-\mathbf{a})$.


## **1.2**
Έστω $\mathbf{A},\mathbf{B},\mathbf{X} \in \mathbb{R}^{d \times d}$ πίνακες με τάξη (rank) ίση με *d*. Να βρείτε, αναλυτικά και σε κλειστή μορφή (closed form solution), το ολικό ελάχιστο του παρακάτω προβλήματος ελαχιστοποίησης:

$\min_{X} \| A - XB \|_F^2$

Να εξηγήσετε, με λεπτομέρια, τη διαδικασία για την εύρεση του ελαχίστου την οποία ακολουθήσατε και να παραθέσετε όλα τα ενδιάμεσα βήματα των πράξεων.

Η Ευκλείδεια νόρμα είναι κυρτή συνάρτηση, συνεχής και παραγωγίσιμη, άρα το σημείο που μηδενίζεται η παράγωγος της θα είναι το ολικό της ελάχιστο, δηλαδή και το ζητούμενο ολικό ελάχιστο του προβλήματος ελαχιστοποίησης.

Αρχικά, θα έχουμε:

$\| A - XB \|_F^2 = tr((A - XB)(A - XB)^H) = $

$tr((A - XB)(A - XB)^T) = $

$tr((A - XB)(A^T - B^TX^T) =$

$tr(AA^T - AB^TX^T - XBA^T + XBB^TX^T) =$

$tr(AA^T) - tr(AB^TX^T) - tr(XBA^T) + tr(XBB^TX^T)$

Για να βρούμε το ολικό ελάχιστο, θα παραγωγίσουμε,

$\frac{\partial}{\partial X}(tr(AA^T) - tr(AB^TX^T) - tr(XBA^T) +tr(XBB^TX^T))=$

$0 - AB^T - (BA^T)^T + (BB^TX^T)^T + XBB^T = 2(XBB^T - AB^T)$

και θα θέσουμε την παραπάνω σχέση ίση με το μηδέν:

$2(XBB^T - AB^T) = 0 \Leftrightarrow XBB^T = AB^T \Leftrightarrow \mathbf{X} = AB^T(BB^T)^{-1}$

Θεωρήστε τον χώρο υποθέσεων που αποτελείται από τις ακόλουθες πραγματικές συναρτήσεις $f_1:\mathbb{R}\to\mathbb{R}$ με δύο παραμέτρους $w=[w_0, w_1]^\intercal\in\mathbb{R}^2$ :

$$
	f_1(x;w) = \frac{1}{1+e^{-(w_0+w_1x)}}
$$

Επιπλέον, τη συνάρτηση κόστους $L_1:\mathbb{R}^2\to ℝ$ η οποία δίνεται ως :

$$
	L_1(w) = -\frac{1}{n} \sum_{i=1}^n \Big( y_i \log_2 f_1(x_i;w) + (1-y_i) \log_2\big( 1 - f_1(x_i;w) \big)	\Big)

Να υλοποιήσετε σε Python και να εκτελέσετε τον αλγόριθμο Gradient Descent ως εξής:

β. ένα (διδιάστατο) γράφημα στο οποίο απεικονίζονται τα δεδομένα εκπαίδευσης (ως σημεία) και το γράφημα του μοντέλου για την τελική εκτίμηση των τιμών των παραμέτρων (ως συνεχής καμπύλη)

```python
# Sigmoid function
def f1(x, w):
    return 1 / (1 + np.exp(-(w[0] + w[1]*x)))

# Cost function
def L1(w, X, y):
    n = len(y)
    pred = f1(X, w)

    return -np.mean(y*np.log2(pred) + (1-y)*np.log2(1 - pred))

# Calculate gradient of cost function
def gradient(X, y, w):
    n = len(y)
    pred = f1(X, w)
    error = pred - y

    # Partial derivatives
    gradient_w0 = np.sum(error) / n
    gradient_w1 = np.sum(error * X) / n

    return np.array([gradient_w0, gradient_w1])

# Gradient descent function
def gradient_descent1(X, y, w_init, learning_rate, T):
    costs = []
    w = np.array(w_init, dtype=float)

    for t in range(T):
        grad = gradient(X, y, w)

        # Update the weights
        w -= learning_rate * grad

        cost = L1(w, X, y)
        costs.append(cost)
        #print(f"Iteration {t}: Cost = {cost}")

    return w, costs
```

```python
w_init = [0., 1.]
learning_rate = 1e-3
T = 150

# Perform gradient descent
w_opt, costs = gradient_descent1(data1_input, data1_output, w_init, learning_rate, T)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot the cost function over iterations
ax1.plot(range(T), costs)
ax1.set_xlabel('Iterations')
ax1.set_ylabel('Cost Function Value')
ax1.grid(True)

# Generate a range of values to plot the logistic regression model
x_values = np.linspace(min(data1_input) - 1, max(data1_input) + 1, 300)

# Calculate probability using the optimal weights
y_values = f1(x_values, w_opt)

# Plot the training data over logistic regression model
ax2.scatter(data1_input, data1_output, color='red', label='Training Data')
ax2.plot(x_values, y_values, color='blue', label='Logistic Regression Model $η=10^{-3}$')
ax2.set_xlabel('Input')
ax2.set_ylabel('Output / Probability')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
```

**α) Παρατητήσεις**: Η γραφική που αποτυπώνει την τιμή της συνάρτησης κόστους έχει καθοδική τάση, άρα καταλαβαίνουμε ότι το κόστος μειώνεται με τις επαναλήψεις. Βλέπουμε ωστόσο, ότι δεν σταθεροποιείται ακόμη, άρα έχει περιθόριο βελτίωσης.

**β) Παρατηρήσεις**: Βλέπουμε πως το μοντέλο μας ταιριάζει καλά στα δεδομένα εκπαίδευσης. Ωστόσο, είναι ορατές και οι αστοχίες για μερικά σημεία.

Να εκτελέσετε εκ νέου τον αλγόριθμο Gradient Descent για κάθε μία από τις παραπάνω περιπτώσεις, χρησιμοποιώντας τις αρχικές τιμές παραμέτρων και τον αριθμό επαναλήψεων που δίνονται στο προηγούμενο ερώτημα, αλλά αυτή τη φορά με μεταβλητό ποσοστό εκμάθησης.

#### **2.2.1**
Χρησιμοποιήστε για τις πρώτες 100 επαναλήψεις $η = 10^{-3}$, για τις επόμενες 40 επαναλήψεις $η=5\cdot10^{-4}$ και τέλος $η=10^{-4}$ για τις τελευταίες 10 επαναλήψεις.

```python
def gradient_descent1_varying_lr(X, y, w_init, learning_rates, iterations):
    w = np.array(w_init, dtype=float)
    costs = []
    total_iterations = sum(iterations)

    # Iterations is a list of numbers
    for i in range(total_iterations):
        if i < iterations[0]:
            learning_rate = learning_rates[0]
        elif i < iterations[0] + iterations[1]:
            learning_rate = learning_rates[1]
        else:
            learning_rate = learning_rates[2]

        grad = gradient(X, y, w)
        w -= learning_rate * grad
        cost = L1(w, X, y)
        costs.append(cost)
        #print(f"Iteration {i}: Cost = {cost}")

    return w, costs
```

```python
w_init = [0., 1.]

# Learning rates and number of iterations for each phase
learning_rates = [1e-3, 5e-4, 1e-4]
iterations = [100, 40, 10]

w_opt_varying_lr, costs = gradient_descent1_varying_lr(data1_input, data1_output, w_init, learning_rates, iterations)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.plot(range(sum(iterations)), costs, label='Varying Learning Rates')
ax1.set_xlabel('Iterations')
ax1.set_ylabel('Cost Function Value')
ax1.grid(True)

x_values = np.linspace(min(data1_input) - 1, max(data1_input) + 1, 300)
y_values_varying_lr = f1(x_values, w_opt_varying_lr)

ax2.scatter(data1_input, data1_output, color='red', label='Training Data')
ax2.plot(x_values, y_values_varying_lr, color='blue', label='Logistic Regression Model (Varying lr)')
ax2.set_xlabel('Input')
ax2.set_ylabel('Output / Probability')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
```

**α) Παρατηρήσεις:** Τώρα το κόστος είναι ελάχιστα μεγαλύτερο από το πρώτο ερώτημα, όμως βλέπουμε την γραφική να σταθεροποιείται προς το ολικό ελάχιστο.

**β) Παρατηρήσεις**: Το μοντέλο μένει σταθερό στην ποιότητα της ακρίβειας του

#### **2.2.2**
Χρησιμοποιήστε αρχικά $η=10^{-3}$ και το οποίο να μειώνετε σε κάθε επόμενη επανάληψη κατά $10^{-6}$ (από αυτό της προηγούμενης της).

```python
def gradient_descent1_decreasing_lr(X, y, w_init, initial_lr, decay, T):
    w = np.array(w_init, dtype=float)
    costs = []

    for i in range(T):
        # Subtract the decay value in every iteration
        learning_rate = initial_lr - i * decay

        grad = gradient(X, y, w)
        w -= learning_rate * grad
        cost = L1(w, X, y)
        costs.append(cost)
        #print(f"Iteration {i}: Cost = {cost:.6f}, Learning Rate = {learning_rate:.6f}")

    return w, costs
```

```python
w_init = [0., 1.]
initial_lr = 1e-3
decay = 1e-6
T = 150

optimized_params, costs = gradient_descent1_decreasing_lr(data1_input, data1_output, w_init, initial_lr, decay, T)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.plot(range(T), costs, label='Decaying Learning Rates')
ax1.set_xlabel('Iterations')
ax1.set_ylabel('Cost Function Value')
ax1.grid(True)

x_values = np.linspace(min(data1_input) - 1, max(data1_input) + 1, 300)
y_values = f1(x_values, optimized_params)

ax2.scatter(data1_input, data1_output, color='red', label='Training Data')
ax2.plot(x_values, y_values, color='blue', label='Logistic Regression Model (Decreasing lr)')
ax2.set_xlabel('Input')
ax2.set_ylabel('Output / Probability')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
```

**α) Παρατηρήσεις**: Η εικόνα που βλέπουμε μοιάζει αρκετά με αυτή του πρώτου ερωτήματος.

**β) Παρατηρήσεις:** Το μοντέλο μένει σταθερό στην ποιότητα ακρίβειας του

**Συμπέρασμα**: Το βέλτιστο αποτέλεσμα προέκυψε από το ερώτημα 2.2.1, όπου το learning rate ξεκινάει σχετικά ψηλά και μειώνεται με βήματα. Αυτό δείχνει πως το μοντέλο κάνει γρήγορη πρόοδο αρχικά και μετά βελτιστοποιεί τις παραμέτρους με πιο σταθερό ρυθμό. $$
	f_2(x;w) = w_0 + w_1 \sigma(w_2+w_3x)
$$
όπου η $\sigma:\mathbb{R}\to\mathbb{R}$ δίνεται ως
$$
	\sigma(x)=\begin{cases} x, & x>0 \\ 0, & x\le 0 \end{cases}
$$

$$
	L_2(w) = \frac{1}{n} \sum_{i=1}^n \big( y_i - f_1(x_i;w) \big)^2
$$

$$
	\frac{d\sigma(\mathbf{x})}{dx} = \begin{cases} 1, & x>0 \\ 0, & x\le 0 \end{cases}

```python
def sigma(x):
    return np.maximum(x, 0)

def sigma_derivative(x):
    return np.where(x > 0, 1, 0)

def f2(x, w):
    return w[0] + w[1] * sigma(w[2] + w[3] * x)

def L2(w, X, y):
    n = len(y)
    pred = f2(X, w)
    # Mean Squared Error
    return np.mean((y - pred)**2)

def gradient_L2(w, X, y):
    n = len(y)
    pred = f2(X, w)
    error = y - pred

    # Calculate gradients for each weight
    gradient_w0 = -2 * np.mean(error)
    gradient_w1 = -2 * np.mean(error * sigma(w[2] + w[3] * X))
    gradient_w2 = -2 * np.mean(error * w[1] * sigma_derivative(w[2] + w[3] * X))
    gradient_w3 = -2 * np.mean(error * w[1] * X * sigma_derivative(w[2] + w[3] * X))

    return np.array([gradient_w0, gradient_w1, gradient_w2, gradient_w3])

def gradient_descent2(X, y, initial_w, learning_rate, iterations):
    w = np.array(initial_w)
    costs = []
    for i in range(iterations):
        cost = L2(w, X, y)
        grad = gradient_L2(w, X, y)
        w -= learning_rate * grad
        costs.append(cost)

    return w, costs
```

#### **2.2.1**
Χρησιμοποιήστε για τις πρώτες 100 επαναλήψεις $η = 10^{-3}$, για τις επόμενες 40 επαναλήψεις $η=5\cdot10^{-4}$ και τέλος $η=10^{-4}$ για τις τελευταίες 10 επαναλήψεις.

```python
def gradient_descent2_varying_lr(X, y, w_init, learning_rates, iterations):
    w = np.array(w_init, dtype=float)
    costs = []
    total_iterations = sum(iterations)

    for i in range(total_iterations):
        if i < iterations[0]:
            learning_rate = learning_rates[0]
        elif i < iterations[0] + iterations[1]:
            learning_rate = learning_rates[1]
        else:
            learning_rate = learning_rates[2]

        cost = L2(w, X, y)
        grad = gradient_L2(w, X, y)
        w -= learning_rate * grad
        costs.append(cost)

    return w, costs
```

#### **2.2.2**
Χρησιμοποιήστε αρχικά $η=10^{-3}$ και το οποίο να μειώνετε σε κάθε επόμενη επανάληψη κατά $10^{-6}$ (από αυτό της προηγούμενης της).

```python
def gradient_descent2_decreasing_lr(X, y, w_init, initial_lr, decay, T):
    w = np.array(w_init, dtype=float)
    costs = []

    for i in range(T):
        learning_rate = initial_lr - i * decay
        grad = gradient_L2(w, X, y)
        w -= learning_rate * grad
        cost = L2(w, X, y)
        costs.append(cost)

    return w, costs
```

## **2.** Να εκπαιδεύσετε την μέθοδο Eigenfaces με d = 30 χρησιμοποιώντας όλες τις εικόνες στο Set_1 (70 εικόνες). Στην συνέχεια χρησιμοποιήστε τα Eigenfaces για να ανακατασκευάσετε μία τυχαία εικόνα του Set_1. Τι παρατηρείτε;

```python
# Implementation of pca using eigenanalysis
def pca(X, n_components):
    # Standardize the data
    X_mean = X - np.mean(X, axis=0)
    X_standardized = X_mean / np.std(X, axis=0)

    cov_matrix = np.cov(X_mean, rowvar=False)

    eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

    # Sort the eigenvalues and corresponding eigenvectors in descending order
    sorted_index = np.argsort(eigenvalues)[::-1]
    sorted_eigenvalues = eigenvalues[sorted_index]
    sorted_eigenvectors = eigenvectors[:, sorted_index]

    principal_components = sorted_eigenvectors[:, :n_components]

    # Project the data onto the new feature space
    projected_data = np.dot(X_mean, principal_components)

    return projected_data, principal_components, np.mean(X, axis=0)

# Image reconstruction function
def reconstruct_image(image, eigenfaces, mean_image):
    projected_image = np.dot(image - mean_image, eigenfaces)
    reconstructed_image = np.dot(projected_image, eigenfaces.T) + mean_image
    return reconstructed_image

set_1, labels = loadImages(faces_path, "Set_1")

projected_data, eigenfaces, mean_image = pca(set_1, 30)

# Choose a random image from Set_1
random_image_index = random.randint(0, len(set_1) - 1)
random_image = set_1[random_image_index]

# Reconstruct the random image
reconstructed_image = reconstruct_image(random_image, eigenfaces, mean_image)

# Reshape the reconstructed image back to its original shape
reconstructed_image = reconstructed_image.reshape(random_image.shape)


# Plot the original and reconstructed random image
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(random_image.reshape((50, 50)), cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(reconstructed_image.reshape((50, 50)), cmap='gray')
axes[1].set_title('Reconstructed Image')
axes[1].axis('off')
plt.show()
```

### **b.** Ποιό d έχει την σταθερότερη απόδοση στα διαφορετικά sets;
